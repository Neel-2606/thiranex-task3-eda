```python
# Exploratory Data Analysis
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load dataset
df = pd.read_csv('../data/student_performance.csv')
df.head()
```

```python
# Basic Information and Statistics
print(df.info())
print('\n\n')
print(df.describe())
```

```python
# Missing Values
print('Missing Values:\n', df.isnull().sum())
```

```python
# Correlation Matrix (Numeric Only)
numeric_df = df.select_dtypes(include=['int64', 'float64'])
plt.figure(figsize=(8,6))
sns.heatmap(numeric_df.corr(), annot=True, cmap='coolwarm', vmin=-1, vmax=1)
plt.title('Correlation Matrix')
plt.show()
```

```python
# Pairplot for trends
sns.pairplot(df, hue='gender')
plt.show()
```

```python
# Distribution of Scores
fig, axes = plt.subplots(1, 3, figsize=(18, 5))
sns.histplot(df['math score'], kde=True, ax=axes[0], color='blue')
axes[0].set_title('Math Score')
sns.histplot(df['reading score'], kde=True, ax=axes[1], color='green')
axes[1].set_title('Reading Score')
sns.histplot(df['writing score'], kde=True, ax=axes[2], color='red')
axes[2].set_title('Writing Score')
plt.show()
```
